```python
%pylab inline
import pandas as pd

df = pd.read_table("GEO_summary.FINAL.txt")
months = """Jan Feb Mar Apr May Jun Jul Aug Sep Oct Nov Dec""".split()
month2num = dict(zip(months, range(13)[1:]))

def getDate(date):
    month, day, year = date.split()
    month = month2num[month]
    year = int(year)
    day = int(day)
    return pd.datetime(year, month, day)

df['date'] = df.series_submission_date.map(getDate)
df = df.sort('date').set_index('date')
df['numPoints'] = df.numProbes * df.numSamples
df['cumSamples'] = df.numSamples.cumsum()
df['cumPoints'] = df.numPoints.cumsum()
```
